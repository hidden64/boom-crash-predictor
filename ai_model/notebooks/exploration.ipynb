# Phase 2 : Exploration IA pour Boom / Crash
J'ai choisi d'utiliser **PyTorch**, car c'est de loin le plus performant et le plus modulable en recherche quantitative pour les séquences.

Ici, tu peux explorer les données récoltées en Phase 1.

```python
import pandas as pd
import matplotlib.pyplot as plt
import os, glob

# 1. Chargement du pool de données de Phase 1
files = glob.glob('../dataset/*.csv')
latest_file = max(files, key=os.path.getctime)
df = pd.read_csv(latest_file)

print(f'Dataset chargé : {len(df)} ticks depuis {latest_file}')
df.tail()
```

### Détection Visuelle des Spikes
Sur les indices Boom (ex: BOOM1000), le prix subit de tout petits drops continus, puis effectue un rebond positif extrême.

```python
plt.figure(figsize=(15, 5))
plt.plot(pd.to_datetime(df['datetime']).iloc[-500:], df['price'].iloc[-500:], color='red')
plt.title('Les 500 derniers ticks du Boom 1000 (Où sont les spikes ?)')
plt.xlabel('Temps')
plt.ylabel('Prix')
plt.grid(True)
plt.show()
```
